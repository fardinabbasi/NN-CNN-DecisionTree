<h1> Q7 </h1>

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
def circular_points(center,radius,n):

    r = np.random.random(n)*(radius[1]-radius[0]) + radius[0]
    alpha= 2 * np.pi * np.random.random(n)
    points = np.vstack((r*np.cos(alpha)+center[0], r*np.sin(alpha)+center[1])).T
    return points

class Adaline:
    def __init__(self, input_size):
        self.weights = np.random.randn(input_size + 1)
        self.z = 0
        
    def predict(self, inputs):
        X = np.insert(inputs, 0, 1)
        self.z = np.dot(self.weights, X)
        return self.activate(self.z)
    
    def train(self, input_vector, target, learning_rate=0.01):
        X = np.insert(input_vector, 0, 1)
        self.weights += learning_rate * (target-self.z)* X
        
    def activate(self, x):
        return 1 if x >= 0 else 0

class Madaline:
    def __init__(self, input_size, hidden_size, learning_rate):
        self.adalines = [Adaline(input_size) for _ in range(hidden_size)]
        self.learning_rate = learning_rate
        
    def train(self, input_vectors, targets, epochs=100):
        for _ in range(epochs):
            for input_vector, target in zip(input_vectors, targets):
                if(self.predict(input_vector)!=target):
                    zs = np.array([np.abs(adaline.z) for adaline in self.adalines])
                    if(target):
                        closest_to0_idx = np.argmin(zs)
                        self.adalines[closest_to0_idx].train(input_vector, 1, self.learning_rate)
                    else:
                        positive_net_idx = np.where(zs>0)
                        for i in positive_net_idx[0]:
                            self.adalines[i].train(input_vector, -1, self.learning_rate)

    def predict(self, input_vector):
        vec = np.array([adaline.predict(input_vector) for adaline in self.adalines])
        return 1 if (np.any(vec)) else 0 # simple logical OR function (MRI)
```

<h2> Part A </h2>

```python
size = 100
circle1 = circular_points([0,0],[0,1],size)
circle2 = circular_points([0,0],[1.3,1.6],size)
plt.figure()
plt.scatter(circle1[:,0],circle1[:,1],c='b',marker='o')
plt.scatter(circle2[:,0],circle2[:,1],c='g',marker='x')
plt.legend(['Class 1','Class 2'])
plt.show()
```

<h2> Part B </h2>

```python
X = np.vstack((circle1,circle2))
Y = np.vstack((np.zeros((size,1)),np.ones((size,1))))
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model1 = Madaline(input_size=2, hidden_size=3, learning_rate=0.05)
model1.train(X_train, y_train)

fig, ax = plt.subplots()
x1 = np.linspace(-2, 2, 100)
for ad in model1.adalines:
    x2 = (ad.weights[1]*x1-ad.weights[0])/ad.weights[2]
    # Plot the lines
    ax.plot(x1, x2, c='black')

ax.scatter(X_test[:,0],X_test[:,1], c=np.where(y_test[:,0]==1, "g", "r"))
ax.set_ylim(-2,2);
```

```python
y_pred = np.array([model1.predict(x) for x in X_test])

print(classification_report(y_test,y_pred))
ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred)).plot();
```

<h2> Part C </h2>

```python
model2 = Madaline(input_size=2, hidden_size=6, learning_rate=0.01)
model2.train(X_train, y_train)

fig, ax = plt.subplots()
x1 = np.linspace(-2, 2, 100)
for ad in model2.adalines:
    x2 = (ad.weights[1]*x1-ad.weights[0])/ad.weights[2]
    # Plot the lines
    ax.plot(x1, x2, c='black')

ax.scatter(X_test[:,0],X_test[:,1], c=np.where(y_test[:,0]==1, "g", "r"))
ax.set_ylim(-2,2);
```

```python
y_pred = np.array([model2.predict(x) for x in X_test])

print(classification_report(y_test,y_pred))
ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred)).plot();
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
```

<h2> Part A </h2>

```python
(X_train, Y_train), (X_test, Y_test) = cifar10.load_data()

for i in np.random.randint(0,len(X_train),5):
    plt.figure()
    plt.imshow(X_train[i])
    plt.axis('off')
    plt.show()
```

<h2> Part B </h2>

```python
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
Y_train = to_categorical(Y_train)
Y_test = to_categorical(Y_test)

x_train, x_valid, y_train, y_valid = train_test_split(X_train, Y_train, test_size=0.8, random_state=42)
```

<h2> Part C & D </h2>

```python
cnn = keras.Sequential(
    [
        layers.Conv2D(32, (3, 3), activation="relu",  input_shape=x_train.shape[1:]),
        layers.Conv2D(32, (3, 3), activation="relu"),
        layers.Conv2D(32, (3, 3), activation="relu"),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        layers.Conv2D(64, (3, 3), activation="relu"),
        layers.Conv2D(64, (3, 3), activation="relu"),
        layers.Conv2D(64, (3, 3), activation="relu"),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        layers.Flatten(),
        layers.Dense(512, activation="relu"),
        layers.Dropout(0.5),
        layers.Dense(10, activation="softmax"),
    ]
)

cnn.summary()
```

<h3> SGD </h3>

```python
cnn.compile(optimizer="sgd",loss="categorical_crossentropy" ,metrics=["Precision"])

history = cnn.fit(x_train, y_train, epochs=10, validation_data=(x_valid, y_valid))
```

```python
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['precision'], label='Training Precision')
plt.xlabel('Epoch')
plt.ylabel('Metric')
plt.title('SGD')
plt.legend()
plt.show()

plt.figure()
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(history.history['val_precision'], label='Validation Precision')
plt.xlabel('Epoch')
plt.title('SGD')
plt.ylabel('Metric')
plt.legend()
plt.show()
```

```python
y_pred_sgd = np.argmax(cnn.predict(X_test), axis=1)

print(classification_report(np.argmax(Y_test, axis=1), y_pred_sgd))
```

<h3> Adam </h3>

```python
cnn.compile(optimizer="adam",loss="mean_squared_error" ,metrics=["Precision"])
history = cnn.fit(x_train, y_train, epochs=10, validation_data=(x_valid, y_valid))
```

```python
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['precision'], label='Training Precision')
plt.xlabel('Epoch')
plt.ylabel('Metric')
plt.title('Adam')
plt.legend()
plt.show()

plt.figure()
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(history.history['val_precision'], label='Validation Precision')
plt.xlabel('Epoch')
plt.title('Adam')
plt.ylabel('Metric')
plt.legend()
plt.show()
```

```python
y_pred_adam = np.argmax(cnn.predict(X_test), axis=1)

print(classification_report(np.argmax(Y_test, axis=1), y_pred_adam))
```

<h3> rmsprop </h3>

```python
cnn.compile(optimizer="rmsprop",loss="mean_squared_error" ,metrics=["Precision"])
history = cnn.fit(x_train, y_train, epochs=10, validation_data=(x_valid, y_valid))
```

```python
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['precision'], label='Training Precision')
plt.xlabel('Epoch')
plt.ylabel('Metric')
plt.title('rmsprop')
plt.legend()
plt.show()

plt.figure()
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(history.history['val_precision'], label='Validation Precision')
plt.xlabel('Epoch')
plt.title('rmsprop')
plt.ylabel('Metric')
plt.legend()
plt.show()
```

```python
y_pred_rmsprop = np.argmax(cnn.predict(X_test), axis=1)

print(classification_report(np.argmax(Y_test, axis=1), y_pred_rmsprop))
```
